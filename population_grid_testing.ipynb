```python
# https://gis.stackexchange.com/questions/78838/converting-projected-coordinates-to-lat-lon-using-python
from pyproj import Transformer
transformer = Transformer.from_crs("EPSG:3035", "EPSG:4326")
# epsg:3035 unit of measurement is 1 meter, so can convert bike station latlons into 3035 coordinates and find distance to center of population grid cells
N = 3482000
E = 3261000
transformed = transformer.transform(N, E)
transformed
```
